#### **Q1** Write a program to swap two numbers using a third variable.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

temp = a
a = b
b = temp
print("After swapping: a =", a, " b =", b)
```

#### **Q2** Write a program to swap two numbers without using a third variable.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

a, b = b, a
print("After swapping: a =", a, " b =", b)
```

#### **Q8** Write a program to check if three numbers form a triangle or not.

#### **Q9** Write a program to check if a triangle is equilateral, isosceles, or scalene.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a==b and b==c:
    print("The triangle is equilateral.")
elif a==b or b==c or a==c: 
    print("The triangle is isosceles.")
else:
    print("The triangle is scalene.")
```

#### **Q14** Write a program to check if a number is single digit, two digit, or three digit.

#### **Q17** Write a program to print the sum of digits of a number.

```python
ip = int(input("Enter a number: "))
sum_digits = 0
while ip > 0:
    sum_digits += ip % 10
    ip //= 10
print("Sum of digits:", sum_digits)
```

#### **Q18** Write a program to count how many digits are present in a number.

```python
ip = int(input())
count = 0

while ip > 0:
    count += 1
    ip //= 10

print("Digits:", count)
```

#### **Q19** Write a program to find the factorial of a number.

```python
n = int(input())
fact = 1

for i in range(1, n+1):
    fact *= i

print("Factorial =", fact)
```

#### **Q20** Write a program to print all prime numbers from 1 to 50.

```python
for num in range(2, 101):
    is_prime = True

    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num)
```
