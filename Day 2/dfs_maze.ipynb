```python
def dfs_any_path(maze, start, goal):
    stack = [start]
    visited = set([start])
    parent = {start: None}
    nodes_explored = 0

    while stack:
        current = stack.pop()
        nodes_explored += 1

        if current == goal:
            # Reconstruct path
            path = []
            node = current
            while node is not None:
                path.append(node)
                node = parent[node]
            return path[::-1], nodes_explored

        r, c = current
        for dr, dc in DIRS:
            nr, nc = r + dr, c + dc
            if in_bounds(maze, nr, nc) and is_walkable(maze, nr, nc):
                if (nr, nc) not in visited:
                    visited.add((nr, nc))
                    parent[(nr, nc)] = current
                    stack.append((nr, nc))

    return None, nodes_explored
```
